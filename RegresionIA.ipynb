Regresión lineal

1. Genere un modelo de regresión lineal simple.
2. Calcule los valores predichos (y^) para todos los valores de x con el modelo de regresión lineal simple.
3. Dibuje los valores reales y los valores predichos en un gráfico de dispersión.
4. Calcule los valores predichos (y^) cuando X = 100 y X = 200.

```python
# Importamos la librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Importamos la función LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
#leemos el documento
data = pd.read_csv('datos_aleatorios.csv')
```

```python
X = data[['X']].values  # La variable independiente
y = data['Y'].values    # La variable dependiente
```

```python
# Creamos un objeto para la regresión lineal
lin_reg = LinearRegression()
#entrenamos los valores
lin_reg.fit(X, y)
```

```python
# Predicciones (opcional, si deseas ver cómo el modelo predice valores)
y_pred = lin_reg.predict(X)
```

```python
# Graficamos los datos y la línea de regresión
plt.scatter(X, y, color='gray', label='Datos originales')
plt.plot(X, y_pred, color='green', label='Línea de regresión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

```python
# Calculamos los valores predichos cuando X = 100 y X = 200
X_test = np.array([[100], [200]])  # Creamos un array con los valores deseados de X
y_test_pred = lin_reg.predict(X_test)
```

```python
# Mostramos los resultados
print(f"Valor predicho para X = 100: {y_test_pred[0]}")
print(f"Valor predicho para X = 200: {y_test_pred[1]}")
```
